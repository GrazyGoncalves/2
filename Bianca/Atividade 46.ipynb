```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cadastro.CSV', encoding='ISO-8859-1', sep=';', low_memory=False)

# Análise dos dados
print("Análise exploratória:")
print("Total de linhas:", len(df))
print("\nResumo estatístico:")
print(df.describe())
print("\nQuantidade de valores NaN por coluna:")
print(df.isna().sum())

# Removendo os nulos
df.dropna(inplace=True)

# Gráfico
state_counts = df['NO_UF_IES'].value_counts()
state_counts.plot(kind='line', title='Contagem de Alunos por Estado', figsize=(10, 6))
plt.xlabel('Estado')
plt.ylabel('Contagem de Alunos')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from fpdf import FPDF

#tratamento de dados
df = pd.read_csv('cadastro.CSV', encoding='ISO-8859-1', sep=';', low_memory=False)
df.head()

#gerar grafico
import plotly.express as px

state_counts = df['NO_UF_IES'].value_counts()
state_counts.plot(kind='line', title='Contagem de Alunos por Estado', figsize=(10, 6))
plt.xlabel('Estado')
plt.ylabel('Contagem de Alunos')
plt.show()

#gerar pdf
from fpdf import FPDF

pdf = FPDF('P', 'mm', 'A4')
pdf.add_page()
pdf.set_font('Arial', '', 16)  # Usando a fonte 'Arial'
pdf.cell(w=0, h=0, txt='Gráfico de Contagem de Alunos por Região', ln=True)


text = 'Porcentagem da flora e fauna ameaçadas no Brasil'
pdf.multi_cell(w=0,h=50, align='J', txt=text)

import os
print(os.getcwd())
pdf.image('contagem.png', x=0, y=20, w=200)
pdf.output('contagem.pdf')
print('pdf salvo')

HUAHUSHUSHSUSHUS
```
